```python
import numpy as np
def 开始():
    a =np.zeros(100)
    z=''
    q=0
    while q==0:
        str = input("请输入：");
        if (str=='结束'):
            break;
        else:
            panduan(str,a)
#用一个while循环来输入，当输入结束的时候停止循环
```

```python
def panduan(str,a):#对最开始语句的判断来执行相应的函数
    s=str.split(" ")
    if (len(s)<=2):
        if (s[0]=='看看'):
            kankanpan(str,a)
        else:
            print("命令不成立！")
    else:
        if (s[2]=='等于'):
            if(s[3].find('负')!=-1):
                print('设置数字不能为负！')
            else:
                bianliang(str,a) 
        elif (s[1]=='增加'):
            if(s[2].find('负')!=-1):
                print('设置数字不能为负！')
            else:
                zengliang(s,a)
        elif (s[1]=='减少'):
            if(s[2].find('负')!=-1):
                print('设置数字不能为负！')
            else:
                zengliang(s,a)
        elif (s[0]=='如果'):
            if(s[3].find('负')!=-1):
                print('设置数字不能为负！')
            else:
                teshupan(str,a)
        else:
            print("命令不成立！")
```

```python
def teshupan(str,a):
    s=str.split(" ")
    global z
    if (z.find(s[1])==-1):#没有创建变量
        print("没有创建相应的变量")
    else:#创建过变量
        teshu(str,a)
```

```python
def teshu(str,a):#自定义判断语句
    s=str.split(" ")
    if (s[2]=='大于'):
        if (len(s[3].split('十'))==1):#判定初始时是否含有'十'
            if(a[z.find(s[1])]>=shuzi(s[3])):
                if (s[5]=='看看'):#当句子中有‘看看’时说明是一种判断语句
                    q=s[6].replace('”','')
                    q=q.replace('“','')
                    print(q)
                else:
                    zengliang(s[5:8],a)
            else:
                if (s[5]=='看看'):
                    q=s[9].replace('”','')
                    q=q.replace('“','')
                    print(q)
                else:
                    ptirnt("无")
        if(len(s[3].split('十'))==2):
            if(a[z.find(s[1])]>=shuzibai(s[3])):
                if (s[5]=='看看'):
                    q=s[6].replace('”','')
                    q=q.replace('“','')
                    print(q)
                else:
                    zengliang(s[5:8],a)
            else:
                if (s[5]=='看看'):
                    q=s[9].replace('”','')
                    q=q.replace('“','')
                    print(q)
                else:
                    ptirnt("无")
    if (s[2]=='小于'):
        if (len(s[3].split('十'))==1):#判定初始时是否含有'十'
            if(a[z.find(s[1])]<shuzi(s[3])):
                if (s[5]=='看看'):
                    q=s[9].replace('”','')
                    q=q.replace('“','')
                    print(q)
                else:
                    zengliang(s[5:8],a)
            else:
                if (s[5]=='看看'):
                    q=s[6].replace('”','')
                    q=q.replace('“','')
                    print(q)
                else:
                    ptirnt("无")
        if(len(s[3].split('十'))==2):
            if(a[z.find(s[1])]<shuzibai(s[3])):
                if (s[5]=='看看'):
                    q=s[9].replace('”','')
                    q=q.replace('“','')
                    print(q)
                else:
                    zengliang(s[5:8],a)
            else:
                if (s[5]=='看看'):
                    q=s[6].replace('”','')
                    q=q.replace('“','')
                    print(q)
                else:
                    ptirnt("无")
```

```python
def bianliang(str,a):#为不同变量寻找相应位置
    s=str.split(" ")
    global z
    if (z.find(s[1])==-1):#没有创建变量
        z=z+s[1]
        chushi(str,a)
    else:#创建过变量
        chushi(str,a)
```

```python
def chushi(str,a):#初始化一个全局变量
    s=str.split(' ')
    if (len(s[3].split('十'))==1):#判定初始时是否含有'十'
        a[z.find(s[1])]=shuzi(s[3])
    if(len(s[3].split('十'))==2):
        a[z.find(s[1])]=shuzibai(s[3])
```

```python
def zengliang(s,a):
    global z
    if (z.find(s[0])==-1):#没有创建变量
        print("没有创建相应的变量")
    else:#创建过变量
        jiajian(s,a)
```

```python
def jiajian(s,a):#对加减函数的调用
    if(fuhao(s,a)<0):
        print('得数为负，错误！')
    else:
        a[z.find(s[0])]=fuhao(s,a)
```

```python
def kankanpan(str,a):
    s=str.split(" ")
    global z
    if (z.find(s[1])==-1):#没有创建变量
        print("没有创建相应的变量")
    else:#创建过变量
        kankan(str,a)
```

```python
def kankan(str,a):#看看 钱包的输出判断
    s=str.split(' ')
    if (a[z.find(s[1])]>20):
        print(hanzibai(a[z.find(s[1])]))
    else:
        print(hanzi(a[z.find(s[1])]))
```

```python
def fuhao(s,a):#增加和减少的运算  这里只有百以内的加减
    if (s[1] == '增加'):
        if (len(s[2].split('十'))==1):#判定初始时是否含有'十'
            return a[z.find(s[0])]+shuzi(s[2])
        if(len(s[2].split('十'))==2):
            return a[z.find(s[0])]+shuzibai(s[2])
    if (s[1] == '减少'):
        if (len(s[2].split('十'))==1):#判定初始时是否含有'十'
            return a[z.find(s[0])]-shuzi(s[2])
        if(len(s[2].split('十'))==2):
            return a[z.find(s[0])]-shuzibai(s[2])
```

```python
def shuzibai(s):#含十的汉字转文字
    if (len(s)!=3):
        if (s == '十'):
            return 10
        if (s == '十一'):
            return 11
        if (s == '十二'):
            return 12
        if (s == '十三'):
            return 13
        if (s == '十四'):
            return 14
        if (s == '十五'):
            return 15
        if (s == '十六'):
            return 16
        if (s == '十七'):
            return 17
        if (s == '十八'):
            return 18
        if (s == '十九'):
            return 19
        if (s == '二十'):
            return 20
        if (s == '三十'):
            return 30
        if (s == '四十'):
            return 40
        if (s == '五十'):
            return 50
        if (s == '六十'):
            return 60
        if (s == '七十'):
            return 70
        if (s == '八十'):
            return 80
        if (s == '九十'):
            return 90
    else:
        a=s.split('十')
        if (a[0] == '二'):
            z=20
        if (a[0] == '三'):
            z=30
        if (a[0] == '四'):
            z=40
        if (a[0] == '五'):
            z=50
        if (a[0] == '六'):
            z=60
        if (a[0] == '七'):
            z=70
        if (a[0] == '八'):
            z=80
        if (a[0] == '九'):
            z=90
        if (a[1] == '一'):
            x=1
        if (a[1] == '二'):
            x=2
        if (a[1] == '三'):
            x=3
        if (a[1] == '四'):
            x=4
        if (a[1] == '五'):
            x=5
        if (a[1] == '六'):
            x=6
        if (a[1] == '七'):
            x=7
        if (a[1] == '八'):
            x=8
        if (a[1] == '九'):
            x=9
        return z+x
```

```python
def shuzi(a):#不含十的汉字转文字
    if (a == '零'):
        return 0
    if (a == '一'):
        return 1
    if (a == '二'):
        return 2
    if (a == '三'):
        return 3
    if (a == '四'):
        return 4
    if (a == '五'):
        return 5
    if (a == '六'):
        return 6
    if (a == '七'):
        return 7
    if (a == '八'):
        return 8
    if (a == '九'):
        return 9
    if (a == '一百'):
        return 100
```

```python
def hanzi(a):#二十以内数字转汉字
    if (a == 0):
        return '零'
    if (a == 1):
        return '一'
    if (a == 2):
        return '二'
    if (a == 3):
        return '三'
    if (a == 4):
        return '四'
    if (a == 5):
        return '五'
    if (a == 6):
        return '六'
    if (a == 7):
        return '七'
    if (a == 8):
        return '八'
    if (a == 9):
        return '九'
    if (a == 10):
        return '十'
    if (a == 11):
        return '十一'
    if (a == 12):
        return '十二'
    if (a == 13):
        return '十三'
    if (a == 14):
        return '十四'
    if (a == 15):
        return '十五'
    if (a == 16):
        return '十六'
    if (a == 17):
        return '十七'
    if (a == 18):
        return '十八'
    if (a == 19):
        return '十九'
    if (a == 20):
        return '二十'
```
